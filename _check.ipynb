```python
import pandas as pd
import folium

csv_file = "/한강다리들.csv"

data = pd.read_csv(csv_file)

m = folium.Map(location=[37.5, 127], zoom_start=10)

for index, row in data.iterrows():
    folium.Marker([row['위도'], row['경도']], popup=row['이름']).add_to(m)

m
```
